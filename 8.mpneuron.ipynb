```python
def AND_gate(x):
    # x is a list of two binary inputs
    return 1 if sum(x) == 2 else 0

def OR_gate(x):
    # x is a list of two binary inputs
    return 1 if sum(x) >= 1 else 0

def NOT_gate(x):
    # x is a single binary input (0 or 1)
    return 1 if x == 0 else 0

# Testing all gates
print("AND Gate:")
for inputs in [[0, 0], [0, 1], [1, 0], [1, 1]]:
    print(f"{inputs} -> {AND_gate(inputs)}")

print("\nOR Gate:")
for inputs in [[0, 0], [0, 1], [1, 0], [1, 1]]:
    print(f"{inputs} -> {OR_gate(inputs)}")

print("\nNOT Gate:")
for input_val in [0, 1]:
    print(f"{input_val} -> {NOT_gate(input_val)}")
```
