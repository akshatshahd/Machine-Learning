```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: size (sqft), price ($1000s)
X = np.array([2104, 1600, 2400, 1416, 3000])
y = np.array([399.9, 329.9, 369.0, 232.0, 539.9])
m = len(X)

# Feature scaling
X = (X - X.mean()) / X.std()

# Initialize weights
w = 0
b = 0
learning_rate = 0.01
epochs = 100

# Gradient descent loop
loss_history = []

for i in range(epochs):
    y_pred = w * X + b
    error = y_pred - y
    loss = (1/(2*m)) * np.sum(error**2)
    loss_history.append(loss)

    # Compute gradients
    dw = (1/m) * np.dot(error, X)
    db = (1/m) * np.sum(error)

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    if i % 10 == 0:
        print(f"Epoch {i}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")

# Final model
print(f"Trained model: price = {w:.2f} * size + {b:.2f}")

# Plotting loss
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss over time")
plt.show()
```
